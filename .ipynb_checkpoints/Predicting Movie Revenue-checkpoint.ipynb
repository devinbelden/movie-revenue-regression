## Explore/Scrub Data

```python
df.isna().sum()
```

### Null Values and Date Errors

All rows with an empty value will be deleted, save for an empty value in `budget`; these values are simply too numerous. Instead, null `budget` values will be replaced with the median budget for that film's year. Median is chosen over mean due to its resistance to skew and outliers, both of which were demonstrated earlier.

Additionally, not all dates are in the correct format. To programmatically change these to `yyyy-mm-dd` format is nigh impossible due to the sheer number of unique datetime formats, so the rows with date errors will be deleted as well.

```python
df_temp = df.copy()
df_temp.dropna(subset=['gross','rating','writer','country','company','runtime'], inplace=True)
df_temp.reset_index(drop=True, inplace=True)
df_temp.isna().sum()
```

```python
df_temp['budget'] = df_temp['budget'].fillna(df_temp.groupby('year')['budget'].transform('median'))
df_temp.isna().sum()
```

```python
date_error = []
dates = df_temp['released'].values
for i in range(len(dates)):
    dates[i] = dates[i][:dates[i].index('(') - 1]
    if len(dates[i].split(' ')) < 3:
        date_error.append(1)
    else:
        date_error.append(0)
        
df_temp['date_error'] = date_error
df_temp = df_temp[df_temp['date_error'] != 1]
df_temp.head()
```

```python
df_temp.drop(['date_error','year'], axis=1, inplace=True)
```

### Reformatting Date

```python
df_temp['release_date'] = df_temp['released'].apply(lambda x: datetime.strptime(x, '%B %d, %Y'))
df_temp['day_of_year'] = df_temp['release_date'].apply(lambda x: x.timetuple().tm_yday)
df_temp['day_of_week'] = df_temp['release_date'].apply(lambda x: x.isoweekday()) #1 is monday, 7 is sunday
df_temp['day_of_year'] = df_temp['day_of_year'].astype(str)
df_temp['day_of_week'] = df_temp['day_of_week'].astype(str)
df_temp.head()
```

```python
df_temp.describe()
```

```python
# Exporting this dataframe for use in Tableau
df_temp.to_csv('imputed_numbers.csv')
```

### Variable Transformations to Handle Skew

```python
df_temp.hist(figsize=(12,12));
```

Since it is a target variable, `gross` will remain untransformed, but `votes`, `budget`, and `runtime` will be transformed. 

`votes` and `runtime` will require a log transform, whereas `budget` does not require such an aggressive transformation. It will receive a cube root transform instead.

```python
df_temp_transform = pd.DataFrame([])

df_temp_transform['votes'] = np.log(df_temp['votes'])
df_temp_transform['runtime'] = np.log(df_temp['runtime'])
df_temp_transform['budget'] = df_temp['budget'] ** (1/3)

df_temp_transform.hist(figsize=(12,12));
```

```python
df_temp.drop(['votes','runtime','budget'], axis=1, inplace=True)
```

```python
df_temp = pd.concat([df_temp, df_temp_transform], axis=1)
df_temp.describe()
```

```python
#for use much later, when transforming human inputs to model-friendly numbers

numerical_stats = {}

for col in list(df_temp.describe().columns):
    numerical_stats[col] = {"Mean": df_temp[col].describe()['mean'],
                            "STD": df_temp[col].describe()['std']}
numerical_stats
```

### Scaling Variables

```python
scaler = StandardScaler()
df_scaled = df_temp.copy()
```

```python
scaling_cols = ['score','votes','runtime','budget']
df_scaled[scaling_cols] = scaler.fit_transform(df_scaled[scaling_cols].values)
df_scaled['scaled_gross'] = scaler.fit_transform(df_scaled['gross'].values.reshape(-1,1))
df_scaled.describe().round(3)
```

```python
df_scaled.corr()
```

```python
for col in scaling_cols + ['scaled_gross']:
    df_scaled[col+'_outliers'] = df_scaled[col].apply(lambda x: abs(x) > 3)

df_scaled = df_scaled[(df_scaled['runtime_outliers'] == False) &
                     (df_scaled['budget_outliers'] == False) &
                     (df_scaled['votes_outliers'] == False) &
                     (df_scaled['score_outliers'] == False) &
                     (df_scaled['scaled_gross_outliers'] == False)]
df_scaled.describe().round(3)
```

```python
len(df_scaled)/len(df)
```

```python
df_scaled.drop(['runtime_outliers','budget_outliers','votes_outliers','score_outliers','scaled_gross_outliers'],
              axis=1, inplace=True)
```

### Checking Categorical Columns

```python
df_scaled[['rating','genre','director','writer','star','country','company']].nunique()
```

There exist far too many unique directors, writers, stars, and companies to employ one-hot encoding. Unfortunately, the machine running this algorithm is not powerful enough to handle that many columns. Instead, any row that does not contain at least one of the top 500 cast and/or crew will not be considered for this model. 

This could, of course, be solved by throwing more power at this problem, which will be done at a later date.

```python
top_500_directors = list(df_scaled.groupby('director')['gross'].median().sort_values(ascending=False).index[:500])
top_500_writers = list(df_scaled.groupby('writer')['gross'].median().sort_values(ascending=False).index[:500])
top_500_stars = list(df_scaled.groupby('star')['gross'].median().sort_values(ascending=False).index[:500])
top_500_companies = list(df_scaled.groupby('company')['gross'].median().sort_values(ascending=False).index[:500])
```

```python
top_500_directors
```

```python
df_filtered = df_scaled[(df_scaled['director'].isin(top_500_directors)) | 
                       (df_scaled['writer'].isin(top_500_writers)) |
                       (df_scaled['star'].isin(top_500_stars)) |
                       (df_scaled['company'].isin(top_500_companies))]
df_filtered.reset_index(drop=True, inplace=True)
print(len(df_filtered))
df_filtered.head()
```

```python
# Exporting this dataset to Tableau
df_filtered.to_csv('data_cleaned_and_scaled.csv')
```

```python
categorical_cols = ['rating','genre','director','writer','star','country','company']
df_categorical = pd.get_dummies(df_filtered, columns=categorical_cols)
df_categorical.head()
```

## Modeling Data

The collection of models chosen for this dataset include: 

* XGBoost
* Gradient Boosting
* Support Vector Machines

Of the three, the highest performing vanilla model will be selected for a Gridsearch for tuning.

```python
X = df_categorical.drop(['name','released','release_date','gross','scaled_gross'], axis=1).values
y = df_categorical['gross'].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=13)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
xgb.fit(X_train,y_train)
```

```python
y_pred = xgb.predict(X_test)
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
```

```python
score
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
```

```python
y_pred = gbr.predict(X_test)
score = r2_score(y_test, y_pred)
```

```python
score
```

```python
from sklearn.svm import SVR

svr = SVR(epsilon=0.2)
svr.fit(X_train, y_train)
```

```python
y_pred = svr.predict(X_test)
score = r2_score(y_test, y_pred)
```

```python
score
```

### Gridsearching for the Best Model

As XGBoost obtained the highest score, it will be selected for a Gridsearch approach.

```python
from sklearn.model_selection import GridSearchCV

go_ahead = input("Cell will take several minutes/hours to run. Do you wish to run this cell (y/n)? ")

if go_ahead == 'y':

    xgb = XGBRegressor()
    grid = {'gamma': [0, 0.3, 0.7],
            'learning_rate': [0.1,0.01],
            'min_child_weight': [3, 5, 10],
            'subsample': [0.7, 0.8, 0.9]}

    gridsearch = GridSearchCV(xgb, param_grid=grid, cv=5)

    xgb_cv = gridsearch.fit(X_train, y_train)
    
    best_params = xgb_cv.best_params_

    print(best_params)
    
else:
    
    best_params = {'gamma': 0,
                   'learning_rate': 0.1,
                   'min_child_weight': 10, 
                   'subsample': 0.8}
    
    print(best_params)
```

```python
xgb_grid = XGBRegressor(**best_params)
xgb_grid.fit(X_train, y_train)

y_pred = xgb_grid.predict(X_test)
score = r2_score(y_test, y_pred)
```

```python
score
```

## Interpretation of the R-squared Value: Feature Importance

According to our XGBoost model, almost 72% of the variance in the data in its current state is explained by the variables within the dataset. The rest of the variance is from something outside the data being looked at. 

Below is a list of the 40 features that were deemed most important within XGBoost's decision making process.

```python
X_columns = list(df_categorical.drop(['name','released','release_date','gross','scaled_gross'], axis=1).columns)
X_columns
```

```python
import operator

rankings = list(zip(xgb_grid.feature_importances_, X_columns))
rankings.sort(key=lambda x: x[0], reverse=True)
rankings[:40]
```

Note that just because a particular column has a high importance does not necessarily mean that it contributed positively towards the model's decision making; it could just as well have contributed to a lower `gross` prediction. 

Below is a graph illustrating the model's residuals (errors in prediction). Since the line separating the two clusters of data is roughly in the middle, that means that our model was wrong in the same magnitude as it was correct. *This is a good thing*. If our model was skewed in one direction or another, that means that the data has an inherent bias that was not accounted for within the data cleaning and transformation stages. 

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.scatter(y_test, y_pred)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)
ax.set_xlabel('Measured revenue')
ax.set_ylabel('Predicted revenue')
plt.title('Measured versus predicted revenue')
plt.ylim((0, 500000000))   # set the ylim to bottom, top
plt.xlim(0, 500000000)     # set the ylim to bottom, top
plt.show()
```

## Predicting New Data

Below is an interactable widget that will feed a user's parameter's into the model constructed within this notebook, and return a prediction for that hypothetical movie's gross. 

```python
import ipywidgets as widgets
from ipywidgets import Dropdown, HBox, Label
```

```python
# Can't do these programmatically, unfortunately...

score_slider = widgets.FloatSlider(
    value=6.4,
    min=0.0,
    max=10.0,
    step=0.1,
    description='Score:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)

day_of_year_slider = widgets.IntSlider(
    value=1,
    min=1,
    max=365,
    step=1,
    description='Day of Year:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

day_of_week_slider = widgets.IntSlider(
    value=1,
    min=1,
    max=7,
    step=1,
    description='Day of Week:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

votes_slider = widgets.IntSlider(
    value=90000,
    min=100,
    max=2500000,
    step=100,
    description='Vote Count:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

runtime_slider = widgets.IntSlider(
    value=107,
    min=60,
    max=380,
    step=10,
    description='Runtime (m):',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

budget_slider = widgets.IntSlider(
    value=32000000,
    min=10000,
    max=350000000,
    step=10000,
    description='Budget:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

sliders = [score_slider, day_of_year_slider, day_of_week_slider, votes_slider, runtime_slider, budget_slider]
```

```python
rating_widget = widgets.Text(
    value='R',
    placeholder='R',
    description='Rating:',
    disabled=False
)

genre_widget = widgets.Text(
    value='Action',
    placeholder='Action',
    description='Genre:',
    disabled=False
)

director_widget = widgets.Text(
    value='Stanley Kubrick',
    placeholder='Stanley Kubrick',
    description='Director:',
    disabled=False
)

writer_widget = widgets.Text(
    value='Rian Johnson',
    placeholder='Rian Johnson',
    description='Writer:',
    disabled=False
)

star_widget = widgets.Text(
    value='Clint Eastwood',
    placeholder='Clint Eastwood',
    description='Star:',
    disabled=False
)

country_widget = widgets.Text(
    value='United States',
    placeholder='United States',
    description='Country:',
    disabled=False
)

company_widget = widgets.Text(
    value='Warner Bros.',
    placeholder='Warner Bros.',
    description='Company:',
    disabled=False
)

text_widgets = [rating_widget, genre_widget, director_widget, writer_widget, star_widget, country_widget, company_widget]
```

```python
def get_prediction(user_input):
    
    """
    Returns a dollar amount that a hypothetical movie with the parameters
    provided by the user would gross. 
    
    Parameters come from the sliders/text boxes in the cell below this 
    function definition.
    """
    
    categorical_prefixes = ['rating_','genre_','director_','writer_','star_','country_','company_']

    numerical_inputs = [i.value for i in user_input[:6]]
    categorical_inputs = [categorical_prefixes[i] + user_input[i+6].value for i in range(len(categorical_prefixes))]

    user_input_df = pd.DataFrame(columns=X_columns)
    user_input_df.loc[0] = 0
    
    for col in categorical_inputs:
        if col in user_input_df.columns:
            user_input_df[col] = 1
        else:
            continue
    
    for i in range(len(numerical_inputs)):
        user_input_df.iloc[:, i] = numerical_inputs[i]
    
    prediction = "{:,}".format(int(xgb_grid.predict(user_input_df.values)[0]))
    
    return "This movie is predicted to gross $" + str(prediction) + "."
```

```python
from ipywidgets import Button, HBox, VBox
title_text = widgets.HTML("<h2 class='text-center'>Movie Revenue Predictor</font><h2>")
display(title_text)
user_inputs = sliders+text_widgets
movie_parameters = VBox(user_inputs)
movie_parameters
```

```python
get_prediction(user_inputs)
```
